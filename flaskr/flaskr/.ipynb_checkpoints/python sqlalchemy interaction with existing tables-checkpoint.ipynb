```python
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker
from sqlalchemy import Column, Integer
```

```python
current_id = 6
```

```python

engine = create_engine('mysql://root:root@localhost/testdb', echo=False)
#engine = create_engine('mysql://root:root@localhost/testdb', echo=True)
Base = declarative_base(engine)
########################################################################
class Entry(Base):
    """"""
    __tablename__ = 'entries'
    __table_args__ = {'autoload':True}
    id = Column(Integer, primary_key=True)
    
    def my_print(self):
        print self.id
        print self.title
        print self.text

def loadSession():
    """"""
    metadata = Base.metadata
    Session = sessionmaker(bind=engine)
    session = Session()
    return session


if __name__ == "__main__":
    session = loadSession()
    
    # insert
    entry = Entry()
    entry.id = current_id
    current_id += 1
    entry.title = 'title2'
    entry.text = 'text2'
    print entry.id
    print entry.title
    print entry.text
    print 'here'
    session.add(entry)
    session.commit()
    
    res = session.query(Entry).all()
    for rec in res:
        if rec is not None:
            rec.my_print()
    
    print '**************query'
    # query
    res = session.query(Entry).filter_by(title='title2')
    for rec in res:
        rec.my_print()
        
    print '**************delete'
    print session.query(Entry).filter_by(title='title2').count()
    res[-1].my_print()
    session.delete(res[-1])
    print session.query(Entry).filter_by(title='title2').count()
    
    print '**************update'
    res[-2].text = "modified"
    session.commit()
    
    ##############
    res = session.query(Entry).all()
    for rec in res:
        if rec is not None:
            rec.my_print()
```
